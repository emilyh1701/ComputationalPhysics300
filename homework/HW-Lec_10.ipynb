Gaussian wavepacket common in physics - short burst of localized wave actions that travels as a unit. Using Fourier Transform, the wave packet can be represented as a superposition of waves with different spatial frequency (like forming a voltage pulse from temporal frequency components). Usually in the form of a function such as: $E(t)=E_0cos(\omega_0t)e^{-t^2/T^2}$. In complex notation: $E(t)=E_0[\frac{e^{i\omega_0t}+e^{-i\omega_0t}}{2}]e^{-t^2/T^2}$. 

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def gpacket(e,t,w0,T):
    return e*np.cos(w0*t)*np.exp(-t**2/T**2) #Our wave packet.

plt.figure(figsize=(10,6))

#define parameters

e=1.0
w0=1.0
T =10.0

npoints = 100
t = np.linspace(-150, 150, npoints)
dt = t[1] - t[0]

sig = gpacket(e,t,w0,T)
trans = np.fft.fftshift(np.fft.fft(sig))*dt/math.sqrt(2*math.pi) 
#trans = np.fft.fft(sig)*dt/math.sqrt(2*math.pi) 
itrans = np.fft.ifftshift(np.fft.ifft(sig))*dt/math.sqrt(2*math.pi)


plt.plot(t, sig, c='k')
plt.plot(t, trans,c='b')
plt.plot(t, itrans,c='g')


plt.show()

print("This illustrates a gaussian wave packet.")
```
